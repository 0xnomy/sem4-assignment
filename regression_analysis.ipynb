```python
#Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score, mean_squared_error
import seaborn as sns
```

**REGRESSION ANALYSIS FOR LINEAR DATA**

```python
#Create linear dataset
np.random.seed(0)  # For reproducibility
x_linear = np.random.rand(100, 1) * 10
y_linear = 2 * x_linear + 3 + np.random.normal(0, 1, (100, 1))

# Fit linear regression
lin_reg = LinearRegression()
lin_reg.fit(x_linear, y_linear)
```

```python
# Linear model coefficients and goodness of fit
lin_coef = lin_reg.coef_
lin_intercept = lin_reg.intercept_
lin_r2 = lin_reg.score(x_linear, y_linear)
lin_rmse = mean_squared_error(y_linear, lin_reg.predict(x_linear), squared=False)
```

```python
print("Linear Regression Results:")
print(f"  Slope: {lin_coef[0][0]}")
print(f"  Intercept: {lin_intercept[0]}")
print(f"  R-Squared: {lin_r2:.4f}")
print(f"  RMSE: {lin_rmse:.4f}")
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=x_linear.ravel(), y=y_linear.ravel(), label='Data')
plt.plot(x_linear, lin_reg.predict(x_linear), color='red', label='Linear Regression')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression")
plt.legend()
plt.show()
```

**REGRESSION ANALYSIS FOR NON-LINEAR DATA**

```python
# Create nonlinear dataset
x_nonlin = np.random.rand(100, 1) * 10
y_nonlin = x_nonlin**2 + 2*x_nonlin + 1 + np.random.normal(0, 1, (100, 1))
# Fit polynomial regression (degree 2)
poly_reg = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
poly_reg.fit(x_nonlin, y_nonlin)
```

```python
# Polynomial model coefficients and goodness of fit
poly_r2 = poly_reg.score(x_nonlin, y_nonlin)
poly_rmse = mean_squared_error(y_nonlin, poly_reg.predict(x_nonlin), squared=False)
print("\nPolynomial Regression Results:")
print(f"  R-Squared: {poly_r2:.4f}")
print(f"  RMSE: {poly_rmse:.4f}")
```

```python
# Plot polynomial regression
plt.figure(figsize=(8, 6))
sns.scatterplot(x=x_nonlin.ravel(), y=y_nonlin.ravel(), label='Data')
x_vals = np.linspace(0, 10, 100).reshape(-1, 1)
plt.plot(x_vals, poly_reg.predict(x_vals), color='red', label='Polynomial Regression')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Polynomial Regression")
plt.legend()
plt.show()
```

# **COMPARISON BETWEEN LINEAR AND NON LINEAR REGRESSION**

**Factors:**

1.   Model performance
2.   Predictive accuracy
3.   Interpretability

**Linear Regression:**
- **R-Squared**: 0.9708
  - This indicates that the linear model explains approximately 97.08% of the variance in the data, suggesting a good fit but potentially leaving some variance unexplained.
- **RMSE (Root Mean Squared Error)**: 0.9962
  - This represents the average error of the linear model's predictions. With an RMSE of 0.9962, the linear model's predictions typically deviate by around 0.9962 units from the actual values.

**Polynomial Regression:**
- **R-Squared**: 0.9993
  - This substantially higher R-Squared value compared to linear regression suggests that the polynomial model captures nearly all (99.93%) of the variance in the data, indicating an excellent fit.
- **RMSE**: 0.9914
  - The slightly lower RMSE of the polynomial model (0.9914) compared to the linear model suggests that, on average, the polynomial model's predictions deviate slightly less from the actual values, indicating better predictive accuracy.

**Interpretability:**

While the polynomial regression model offers superior predictive performance, it may sacrifice some interpretability due to its more complex nature compared to the linear regression model. The linear model's coefficients directly represent the relationship between variables, making it easier to interpret, whereas the polynomial model involves higher-order terms, which can be more challenging to interpret intuitively.

**Conclusion**

The data suggests that polynomial regression outperforms linear regression in capturing the variance and predicting outcomes, indicating a likely nonlinear relationship between variables. The polynomial model's higher R-Squared and slightly lower RMSE imply better predictive accuracy.
